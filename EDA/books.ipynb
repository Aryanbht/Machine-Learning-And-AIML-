```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
df.head()
```

```python
df.describe()
```

```python
df = df[df['Publishing Year']  > 1900 ]
```

```python
df.isna().sum()
```

```python
df.dropna(subset='Book Name' , inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number Of Books')
plt.title('Number Of Books By Genre')
```

```python
df.groupby('Author')['Book_average_rating'].mean()
```

```python
sns.boxplot(x = 'genre' , y = 'Book_ratings_count' , data= df)
plt.xlabel('genre')
plt.ylabel('Rating Of Books ')
plt.title('Book Rating Count of eac genre ')
```

```python
plt.scatter(df['sale price'] , df['units sold'])
plt.xlabel('Sale Price')
plt.ylabel('Unit Sold')
plt.title('Scatter Plot Of Sales Price vs Unit Sold')
```

```python
lang_counts  = df['language_code'].value_counts().sort_values(ascending=True)
```

```python
plt.pie(lang_counts, labels=lang_counts.index , startangle=90 )
plt.show()
```
